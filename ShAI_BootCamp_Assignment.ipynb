# Data Cleaning

- drop Notes and Status columns as they do not have any values
- drop Agency column as it has only one value
- All the Nan values in the Benefits, OtherPay and OvertimePay columns  will be replaced with 0 as a way they do not take any benefits
- Nan values in BasePay will be replaced with the mean value of BasePay grouped with JobTitle

```python
df.drop(columns = ['Notes','Status','Agency'],inplace = True)
```

```python
df['Benefits']=df['Benefits'].fillna(0)
df['OtherPay']=df['OtherPay'].fillna(0)
df['OvertimePay']=df['OvertimePay'].fillna(0)
```

```python
job_base= df.groupby('JobTitle')['BasePay'].agg('mean').to_dict()
```

```python
df['BasePay'] = df.apply(lambda row : job_base[row['JobTitle']] if pd.isnull(row['BasePay']) else row['BasePay'],axis = 1)
```

```python
df.isna().sum()
```

The remaing 8 rows will be droped as they are unique jobtitles

```python
df.dropna(inplace=True)
```
